alpha diversity analysis to GPGC data using IGS method
========


For alpha diversity, we can generate a list of IGSs and the respective abundance in a sample. Then existing estimators like Chao's can be used to estimate total number of IGSs in the sample. Rarefaction curve based on number of IGSs can also be genereated.

For more information about IGS-based method, pleaes see:
http://nbviewer.ipython.org/github/qingpeng/2013-diversity/blob/master/notebook/2013-diversity-summary.ipynb


It needs further work to do the analysis to the whole GPGC data sets, here only 1 million reads and 2 million reads are used seperately. But we can still see some interesting results here.

WAit wait!!
======

We didn't take the variable length of the reads of different samples into account!

The reads lengths of the 8 samples are not the same. 


- Iowa corn: 100bp          
- Iowa prairie: 100bp
- wisconsin_restore: 151bp
- all other: 114bp

k = 20

The length of IGSs is related to the read length.

```
Length_of_IGS = Length_of_read - size_of_k +1
```

So if we want to estimate the size of metagenome (novel information in a sample) from number of IGSs, we should take the read length into account.

Next figure is about the observed size of metagenome, and chao1-estimated size of metagenome.

Conclusion for now:
----

1. As shown in the test to other data set, the IGS-based alpha analysis can compare the richness of samples pretty well. It is safe to say some samples have a higher richness than other samples. But it is still difficut to estimate the absolute value of the richness.

2. Only using a really small proportion of data set, like 1 million reads, we can see clear separation of samples. How reliable is this? It's worth further investigation.

3. Actually just for the purpose of comparing the richness of different samples, counting unique k-mers or counting non-redundant reads as we did before should have similar results. I haven't compared the performance specifically. Theoretically IGS-based method should have a better accuracy than k-mer counting. Also, this method supplies a more straightfoward and clean framework to do alpha diversity analysis. With this framework, we can try any other estimators, besides chao1 to estimate richness. Estimation of evenness is also possible. 

4. To solve the problem and make better estimation, there are several potential solutions:
    - Remove/Correct error as much as possible. The sequencing error is still daunting, causing the abundance of IGSs inaccurate, especially the rarer ones, which are important to chao1 estimator.
    - Try more estimators to find more appropriate one, Chao1 estimator only looks at rarer units with abundance of 1 and 2. Other estimator may work better. 
    - Technical problem. It is too slow to use QIIME to deal with the IGS abundance table. I need to adjust the parameters of QIIME more carefully like reducing random subsampling cycle or write some codes to deal with IGS table more efficiently. I think there may be different way to treat big IGS table from OTU table.

````python
read_length={}
read_length['iowa_corn'] = 100
read_length['iowa_prairie'] = 100
read_length['kansas_corn'] = 114
read_length['kansas_prairie'] = 114
read_length['wisconsin_corn'] = 114
read_length['wisconsin_prairie'] = 114
read_length['wisconsin_restored'] = 151
read_length['wisconsin_switchgrass'] = 114

import pandas as pd
import numpy as np
import scipy
from IPython.display import HTML
from scipy.cluster.hierarchy import linkage, dendrogram
from scipy.spatial.distance import pdist, squareform
from IPython.display import Image
from pandas import DataFrame, read_csv, Series
def alpha_plot_confg(alpha_file, readlengths): 
    # alpha_file is result file generated by QIIME, if all samples have same readlength, readlength is a integer, if 
    # samples have variable readlength, readlength is a dictionary
    df = read_csv(alpha_file,sep="\t",na_values=["n/a"])

    # remove first column
    df2 = df.ix[:,1:]
    df3 = df2.astype(np.float64)
    grouped = df3.groupby('sequences per sample')
    df_mean = grouped.mean().drop('iteration',axis=1)
    

    # variable readlength
    if type(readlengths) == type({}):
        #print "dictionary"
        df_mean2=df_mean.copy()
        for column in df_mean:
            #print column
           # print frame
            df_mean2[column] = df_mean[column] * (readlengths[column]-20+1)
    else: # for same readlength
        df_mean2=df_mean.copy()
        for column in df_mean:
            #print column
           # print frame
            df_mean2[column] = df_mean[column] * (readlengths-20+1)
    #print df_mean2.index
    df_mean2 = df_mean2.drop(df_mean2.index[-1]) # remove line with na
    #print df_mean2
    # build multi-index data frame for groupby
    df_mean3 = DataFrame(np.arange(80).reshape((10,8)), index=df_mean2.index, columns = [[x.split('_')[0] for x in df_mean.columns],[x.split('_')[1] for x in df_mean2.columns]])
    df_mean3.columns.names = ['location','treatment']
    for column in df_mean2:
        location,treatment = column.split('_')
        df_mean3[location,treatment] = df_mean2[column]
    location_mean = df_mean3.groupby(level='location',axis=1).mean()
    location_std = df_mean3.groupby(level='location',axis=1).std()
    treatment_mean=df_mean3.groupby(level='treatment',axis=1).mean()
    treatment_std=df_mean3.groupby(level='treatment',axis=1).std()
    plot_fig = (df_mean3, location_mean, location_std, treatment_mean, treatment_std)
    return plot_fig


    
    
def alpha_plot(chao_file,obs_file,readlength, figtitle):
    plot_fig_chao = alpha_plot_confg(chao_file,readlength)
    plot_fig_obs = alpha_plot_confg(obs_file,readlength)
    (df_mean3, location_mean, location_std, treatment_mean, treatment_std) = plot_fig_chao
    fig, ax = plt.subplots(figsize=(16,10),nrows=2, ncols=2)
    fig.suptitle (figtitle,fontsize=20)
    
    df_mean3.plot( ax=ax[0,1],kind='line', linewidth=2.0, title = "Figure 2: chao1 estimated, by sample")
    treatment_mean.plot(yerr=treatment_std, ax=ax[1,1], kind='line', linewidth=2.0, title = "Figure 4: chao1 estimated, by treatment ")
    location_mean.plot(yerr=location_std, ax=ax[1,0], kind='line', linewidth=2.0, title = "Figure 3: chao1 estimated, by location ")
    
    (df_mean3_obs,a ,b,c,d) = plot_fig_obs
    df_mean3_obs.plot(yerr=location_std, ax=ax[0,0],kind='line', linewidth=2.0, title = "Figure 1: observed")
````
