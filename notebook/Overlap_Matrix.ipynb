```python
def read_matrix(filename):
    
    file_obj = open(filename,'r')
    dimension = file_obj.readlines()[-1].split()[0].split('-')
    #print dimension
    L1 = []
    L2 = []
    L3 = []
    for x in range(int(dimension[0])+1):
        L1.append([0]*(int(dimension[1])+1))
        L2.append([0]*(int(dimension[1])+1))
        L3.append([0]*(int(dimension[1])+1))
    file_obj = open(filename,'r')
    for line in file_obj:
        line = line.rstrip()
        fields = line.split()
        #print test2
        #print fields
        L1[int(fields[0].split('-')[0])][int(fields[0].split('-')[1])] = int(fields[1])
        L2[int(fields[0].split('-')[0])][int(fields[0].split('-')[1])] = int(fields[2])
        L3[int(fields[0].split('-')[0])][int(fields[0].split('-')[1])] = int(fields[3])
    file_obj.close()
    #print "test"
    #print L1
    #print L2
    #print L3
    #print dimension
    
    return (dimension,L1,L2,L3)
    
```

```python
def draw_figure(filename,title,dimension1,dimension2):
        
    #print column_labels
    (dimension,L1,L2,L3) = read_matrix(filename)
    #print L1
    #print L2
    #print L3
    if dimension1 !=0 or dimension2 !=0:
        dimension = [dimension1,dimension2]
    print dimension
    figure,((ax1,ax2),(ax3,ax4)) = plt.subplots(2,2)
    #figure.set_title('test1')
    ax1.set_title('reads in 1st sample',size=5)
    ax2.set_title('reads in 2nd sample',size=5)
    ax3.set_title('reads in both samples',size=5)
    ax = [ax1,ax2,ax3]
    L = [L1,L2,L3]
    for n in range(3):
        draw_heatmap(dimension,ax[n],L[n])
    #plt.show()
    fig = plt.gcf()

    fig.suptitle(title)
```
