A series of synthetic data sets were generated.

* 1M genome 1X coverage
* 1M genome 5X
* 1M genome 10X
* 1M genome 20X
* 1M genome 40X
* combine three data sets above (1X, 10X, 40X)
* combine all 5 data sets above (1X, 5X, 10X, 20X, 40X)
* combined reads with 5 species with abundance as (1X:2X:3X:4X:5X)
* combined reads with 5 species with abundance as (1X:2X:4X:8X:16X)

All With illumina error pattern


```python
saturation1 = numpy.loadtxt('../data/evenness_1-Empirical.fna.gz.C1.report')
saturation10 = numpy.loadtxt('../data/evenness_10-Empirical.fna.gz.C1.report')
saturation40 = numpy.loadtxt('../data/evenness_40-Empirical.fna.gz.C1.report')
saturation1_10_40 = numpy.loadtxt('../data/evenness_1_10_40-Empirical.fna.gz.C1.report')
saturation5 = numpy.loadtxt('../data/evenness_5-Empirical.fna.gz.C1.report')
saturation20 = numpy.loadtxt('../data/evenness_20-Empirical.fna.gz.C1.report')
saturation_1_5_10_20_40 = numpy.loadtxt('../data/evenness_1_5_10_20_40-Empirical.fna.gz.C1.report')
saturation12345 = numpy.loadtxt('../data/evenness_12345-Empirical.fna.gz.C1.report')
saturation124816 = numpy.loadtxt('../data/evenness_124816-Empirical.fna.gz.C1.report')
```

Saturation of 10X data: 
    Y-axis is the fraction of non-redundant reads (coverage<1, this means, the fraction of reads out of all loaded reads[120000] that do not have overlap with other reads)

```python
plot(saturation10[:,0], saturation10[:,1] / max(saturation10[:,0]))
axis(xmin=0, ymin=0)
title('10X coverage')
xlabel('number of reads loaded')
ylabel('fraction of non-redundant reads (coverage=1)')
```

Meaning of this figure!!!!
What is x-axis and y-axis????



Let's check the number. Exactly the same shape. But it seems that the number of non-redundant continues increasing as we load more reads


```python
plot(saturation10[:,0], saturation10[:,1])
axis(xmin=0, ymin=0)
title('10X coverage')
xlabel('number of reads loaded')
ylabel('number of non-redundant reads (coverage=1)')
```

If we set the y-axis max as the number of all the loaded reads. We can say it is close to saturation.

```python
plot(saturation10[:,0], saturation10[:,1])
axis(xmin=0, ymin=0,ymax=max(saturation10[:,0]))
title('10X coverage')
xlabel('number of reads loaded')
ylabel('number of non-redundant reads (coverage=1)')
```
