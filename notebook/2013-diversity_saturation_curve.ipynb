# Saturation curve of IGSs for simulated sequencing reads of e.coli (with error correction!)

##Error correction does work!

The curve of 1% error rate is perfect now. It looks like the influence of sequencing error is dramatically eliminated after such error checking method. After error correction, the remaining errors will no longer pose a threat to the median kmer counting.

For 2% error rate, after error correction, the curve is better than that of 1% before correction.

The Chao1 estimators works so good after error correction, too good to be true. I will look into this.



```python
import pandas as pd
import numpy as np
import scipy
from IPython.display import HTML
from scipy.cluster.hierarchy import linkage, dendrogram
from scipy.spatial.distance import pdist, squareform
from IPython.display import Image
from pandas import DataFrame, read_csv
```

```python
def make_df(file_name1,file_name2):

    df1 = read_csv(file_name1,sep="\t")
    df2 = read_csv(file_name2,sep="\t")
    df1_p = df1.loc[:,['sampleI','sampleJ']]
    df1_p.index = df1["sequences per sample"].tolist()
    df2_p = df2.loc[:,['sampleB','sampleC','sampleD']]
    df2_p.index = df2["sequences per sample"].tolist()
    df12 = df2_p.join(df1_p)
    df12['T'] = 57279
    #print(df12)
    df12.columns=['no error','1% error','2% error','1% error_corrected','2% error_corrected','expected #']
    return df12
```

```python
file_name1 = "../data/observed_species_correct.txt"
file_name2 = "../data/observed_species_original.txt"

result = make_df(file_name1,file_name2)
result
df = result.iloc[:,0:5]
expect = result['expected #']
```

```python
ax = df.plot(figsize=(10, 8),linewidth=3, title="observed_IGS", xlim =(0,500000))
ax.set_xlabel('# of sampled reads')
ax.set_ylabel('# of observed IGS')
expect2=expect.append(pd.Series({500000:57279}))
#print expect2
expect2.plot(style = 'r--',label='expected # of IGSs',linewidth=1.5)
```

```python
file_name1 = "../data/chao1_correct.txt"
file_name2 = "../data/chao1_original.txt"


result = make_df(file_name1,file_name2)
result
df = result.iloc[:,0:5]
expect = result['expected #']
```

```python
ax = df.plot(figsize=(10, 8),linewidth=3, title="chao1", ylim =(0,350000))
ax.set_xlabel('# of sampled reads')
ax.set_ylabel('chao1')
expect2.plot(style = 'r--',label='expected # of IGSs',linewidth=1.5)
```
