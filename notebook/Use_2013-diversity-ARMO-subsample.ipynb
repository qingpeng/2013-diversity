```python
from skbio.diversity.beta import pw_distances
from skbio import DistanceMatrix
from pandas import DataFrame, read_csv,Series
from skbio.stats.distance import mantel
import matplotlib.pyplot as plt
```

```python
DIR = "/Users/qingpeng/GDrive/Development/Github/2013-diversity/data/ARMO/Subsample/"
```

```python
real_df = read_csv(DIR+'bdiv_even1000-8M/bray_curtis_dm.txt',sep="\t", header = 0,index_col=0)
real_df
```

```python
df_4M = read_csv(DIR+'bdiv_even1000-4M/bray_curtis_dm.txt',sep="\t", header = 0,index_col=0)
df_1M = read_csv(DIR+'bdiv_even1000-1M/bray_curtis_dm.txt',sep="\t", header = 0,index_col=0)
```

```python
df_2M_0 = read_csv(DIR+'bdiv_even1000-2M-0/bray_curtis_dm.txt',sep="\t", header = 0,index_col=0)
df_2M_1 = read_csv(DIR+'bdiv_even1000-2M-1/bray_curtis_dm.txt',sep="\t", header = 0,index_col=0)
df_2M_2 = read_csv(DIR+'bdiv_even1000-2M-2/bray_curtis_dm.txt',sep="\t", header = 0,index_col=0)
```

```python
correlation = {}
```

```python
correlation["4M"] = mantel(real_df,df_4M,method='spearman', strict=False)[0]
```

```python
correlation["2M-0"] = mantel(real_df,df_2M_0,method='spearman', strict=False)[0]
correlation["2M-1"] = mantel(real_df,df_2M_1,method='spearman', strict=False)[0]
correlation["2M-2"] = mantel(real_df,df_2M_2,method='spearman', strict=False)[0]
#correlation["1M"] = mantel(real_df,df_1M,method='spearman', strict=False)[0]
```

```python
import numpy as np
import matplotlib.pyplot as plt
ind = np.arange(4)  # the x locations for the groups
width = 0.6       # the width of the bars

fig, ax = plt.subplots(figsize=(10, 8))
rects1 = ax.bar(ind+0.3, correlation.values(), width, color='b')


# add some text for labels, title and axes ticks
ax.set_ylabel('Mantel Correlation', fontsize =25)
ax.set_xlabel('Subsample with different size', fontsize =25)
ax.set_title('Accuracy of similarity matrix calculated using subsamples', fontsize =25)
ax.set_xticks(ind+width)
ax.set_xticklabels( correlation.keys(), fontsize =25 )
ax.set_ylim(0.97,1.0)
savefig('mental_subsample.pdf')
```
