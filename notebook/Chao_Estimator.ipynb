```python
read_coverage1 = numpy.loadtxt('../data/evenness_1-Empirical.fna.gz.hist')
read_coverage10 = numpy.loadtxt('../data/evenness_10-Empirical.fna.gz.hist')
read_coverage40 = numpy.loadtxt('../data/evenness_40-Empirical.fna.gz.hist')
read_coverage1_10_40 = numpy.loadtxt('../data/evenness_1_10_40-Empirical.fna.gz.hist')
read_coverage5 = numpy.loadtxt('../data/evenness_5-Empirical.fna.gz.hist')
read_coverage20 = numpy.loadtxt('../data/evenness_20-Empirical.fna.gz.hist')
read_coverage_1_5_10_20_40 = numpy.loadtxt('../data/evenness_1_5_10_20_40-Empirical.fna.gz.hist')
read_coverage12345 = numpy.loadtxt('../data/evenness_12345-Empirical.fna.gz.hist')
read_coverage124816 = numpy.loadtxt('../data/evenness_124816-Empirical.fna.gz.hist')



```

```python
def chao(read_coverage):
    s_obs = sum(read_coverage[:,1])
    s_chao = s_obs + read_coverage[1,1]**2/(2*(read_coverage[2,1]/2))
    #print int(s_chao)
    #print unique_reads(read_coverage)
    return (int(sum(unique_reads(read_coverage))),int(s_chao))

def unique_reads(read_coverage):
    unique_read = []
#print read_coverage40[:,0]
    for i in range(1,len(read_coverage[:,0])):
 #   print i
        n = read_coverage[i,1]/read_coverage[i,0]
        unique_read.append(n)
  #  print n
    return unique_read
#print unique_reads
```

```python
plot(read_coverage1[:,0], read_coverage1[:,1],'bo-',label='1x')
plot(read_coverage10[:,0], read_coverage10[:,1],'go-',label='10x')
plot(read_coverage40[:,0], read_coverage40[:,1],'ro-',label='40x')
plot(read_coverage1_10_40[1:,0], read_coverage1_10_40[1:,1],'k-',label='1X,10X,40X combined')
xlabel('Estimated sequencing depth')
ylabel('Number of reads with that depth')
axis([0,70,0,60000])
legend()
savefig('igs3.pdf',dpi=300)
```

The y-axis is the number of reads with a specific coverage. So it's like 40000 reads have coverage as 25. In fact we can say that there are about 40000/25 = 1600 distinct segments. Here the segments can be considered as the segments in genome. Let's divide the number of reads by the coverage and draw a plot.

```python
plot(read_coverage1[1:,0], unique_reads(read_coverage1),'bo-',label='1x')
plot(read_coverage10[1:,0], unique_reads(read_coverage10),'go-',label='10x')
plot(read_coverage40[1:,0], unique_reads(read_coverage40),'ro-',label='40x')
plot(read_coverage1_10_40[1:,0], unique_reads(read_coverage1_10_40),'k-',label='1X,10X,40X combined')
xlabel('Estimated sequencing depth')
ylabel('Number of IGSs with that depth')
axis([0,50,0,6000])
legend()
savefig('igs4.pdf',dpi=300)
print '# of observed 10X unique segments: (without coverage=1):',sum(unique_reads(read_coverage10)[1:])
print '# of observed 40X unique segments: (without coverage=1):',sum(unique_reads(read_coverage40)[1:])
print '# of observed 10X unique segments: (with coverage=1):',sum(unique_reads(read_coverage10)[:])
print '# of observed 40X unique segments: (with coverage=1):',sum(unique_reads(read_coverage40)[:])
print '# of observed 1-1040-X unique segments: (without coverage=1):',sum(unique_reads(read_coverage1_10_40)[1:])
print (28784.4805618-19558.4805618)/6+19558.4805618
print (54721.9929483-21948.9929483)/23+21948.9929483
```

So the area below each peak for 10x, 40x should be similar in the figure, since the number of distinct segments in a genome is fixed. 

To calculate Chao estimator, we should use the number of segments, rather than the reads.

Another concern is the segment with coverage as 1. Many of them are due to sequencing error. 

So Chao's estimator may not be the best choice. Chao's estimator uses the number of unique elements and elements appearing twice. 

Other estimator use more numbers rather than only the number of unique elements.


```python
s = chao(read_coverage1)
print "               1M genome 1X",s[0],s[1]
s = chao(read_coverage5)
print "               1M genome 5X",s[0],s[1]
s = chao(read_coverage10)
print "              1M genome 10X",s[0],s[1]
s = chao(read_coverage20)
print "              1M genome 20X",s[0],s[1]
s = chao(read_coverage40)
print "              1M genome 40X",s[0],s[1]
s = chao(read_coverage1_10_40)
print "       1M genome 1X:10X:40X",s[0],s[1]
s = chao(read_coverage_1_5_10_20_40)
print "1M genome 1X:5X:10X:20X:40X",s[0],s[1]
s = chao(read_coverage12345)
print "   1M genome 1X:2X:3X:4X:5X",s[0],s[1]
s = chao(read_coverage124816)
print "  1M genome 1X:2X:4X:8X:16X",s[0],s[1]
```
