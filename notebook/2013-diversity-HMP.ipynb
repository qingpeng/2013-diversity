```python
def scatter_3d(ord_results, df, column, color_map, title='', axis1=0,
...                axis2=1, axis3=2):
...    coord_matrix = ord_results.site.T
...    ids = ord_results.site_ids
...    colors = [color_map[df[column][id_]] for id_ in ord_results.site_ids]
...
...    fig = plt.figure(figsize=(8, 6))
...    ax = fig.add_subplot(111, projection='3d')
...
...    xs = coord_matrix[axis1]
...    ys = coord_matrix[axis2]
...    zs = coord_matrix[axis3]
...    plot = ax.scatter(xs, ys, zs, c=colors,s=40)
...
...    ax.set_xlabel('PC %d' % (axis1 + 1))
...    ax.set_ylabel('PC %d' % (axis2 + 1))
...    ax.set_zlabel('PC %d' % (axis3 + 1))
...    ax.set_xticklabels([])
...    ax.set_yticklabels([])
...    ax.set_zticklabels([])
...    ax.set_title(title)
...    return fig
```

```python

def pcoa_plot(dm_file, map_file):

    dm = DistanceMatrix.from_file(dm_file)
    df = pd.read_csv(map_file, sep='\t', index_col=0)
    bc_pc = PCoA(dm).scores()
    coord_matrix = bc_pc.site.T
    #coord_matrix
    pro = bc_pc.proportion_explained 
    fig = scatter_3d(bc_pc, df, 'tissue', {'throat': 'g', 'buccal_mucosa': 'b','anterior_nares': 'r','posterior_fornix':'y'},
                      'Samples colored body part')
    anosim = ANOSIM(dm, df, column='tissue')
    results = anosim(999)
    R = results.statistic
    p = results.p_value
    print "R-value:",R
    print "p-value:",p
```
